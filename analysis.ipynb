## Encoding the Target Variable

Machine learning algorithms require the target variable to be represented in numerical form rather than text. Since the **Attrition** column contains categorical values (**Yes** and **No**), it is converted into numerical labels before model development.

In this project, **Yes** is encoded as **1** (Employee Left) and **No** is encoded as **0** (Employee Stayed). This transformation enables classification algorithms to learn meaningful patterns from the dataset.

```python
# ============================================================
# Encoding the Target Variable
# ============================================================

# Converting Attrition into numerical values
data["Attrition"] = data["Attrition"].map({
    "Yes": 1,
    "No": 0
})

print("Target variable encoded successfully.\n")

print(data["Attrition"].value_counts())
```

### Observation

- The target variable **Attrition** has been successfully converted from categorical values (**Yes/No**) to numerical values (**1/0**).
- Employee attrition is represented by **1**, while employee retention is represented by **0**.
- This transformation makes the target variable compatible with machine learning classification algorithms.

### Business Interpretation

Converting the target variable into numerical form enables machine learning models to accurately distinguish between employees who are likely to leave and those who are expected to remain with the organization. This transformation is an essential preprocessing step for building reliable employee attrition prediction models.

## Encoding Categorical Features

Most machine learning algorithms cannot directly process categorical (text-based) variables. Therefore, categorical features must be transformed into numerical representations before model training.

This project uses **One-Hot Encoding**, a widely adopted preprocessing technique that converts each category into a separate binary feature while preserving the original information. Since the target variable **Attrition** has already been encoded, only the remaining categorical input features are transformed.

```python
# ============================================================
# Identifying Categorical Features for Encoding
# ============================================================

categorical_columns = data.select_dtypes(include=["object"]).columns

print("Categorical Features to be Encoded:\n")
print(list(categorical_columns))
```

### Observation

- The dataset contains **7 categorical input features** that require numerical transformation before model development.
- The target variable **Attrition** is no longer included because it has already been successfully encoded into numerical values.
- These categorical features will be converted into numerical format using **One-Hot Encoding**, ensuring compatibility with machine learning classification algorithms.

---

## Applying One-Hot Encoding

After identifying the categorical features, One-Hot Encoding is applied to transform each categorical variable into binary indicator variables. This preprocessing technique enables machine learning algorithms to effectively utilize categorical information without introducing ordinal relationships between categories.

```python
# ============================================================
# Applying One-Hot Encoding
# ============================================================

data = pd.get_dummies(
    data,
    columns=categorical_columns,
    drop_first=True,
    dtype=int
)

print("One-Hot Encoding completed successfully.\n")

print("Dataset Shape After Encoding :", data.shape)
```

### Observation

- One-Hot Encoding was successfully applied to all **7 categorical input features**.
- The number of features increased from **31 to 45**, indicating that categorical variables were transformed into binary indicator variables.
- The number of employee records remained unchanged (**1,470**), confirming that only the feature representation changed while preserving the original data.

### Business Interpretation

One-Hot Encoding preserves valuable employee information while converting categorical attributes into a format that machine learning algorithms can understand. This enables the predictive models to effectively learn relationships between employee characteristics and attrition without introducing artificial ranking among categories.

After completing data preprocessing, the dataset is prepared for machine learning model development. This stage involves separating the input features from the target variable and splitting the dataset into training and testing sets.

## Separating Features and Target Variable

Machine learning models require the input variables (features) and the output variable (target) to be separated before training. The feature matrix contains all employee-related attributes, while the target variable represents employee attrition.

```python
# ============================================================
# Separating Features and Target Variable
# ============================================================

X = data.drop("Attrition", axis=1)

y = data["Attrition"]

print("Feature Matrix Shape :", X.shape)
print("Target Variable Shape :", y.shape)
```

## Splitting the Dataset into Training and Testing Sets

To evaluate model performance fairly, the dataset is divided into training and testing subsets. The training dataset is used to learn patterns from employee information, while the testing dataset is reserved for evaluating how well the trained model performs on unseen employee records.

An **80:20 split** is used in this project, which is a widely accepted practice for supervised machine learning tasks.

```python
# ============================================================
# Splitting the Dataset
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Feature Shape :", X_train.shape)
print("Testing Feature Shape  :", X_test.shape)

print("\nTraining Target Shape :", y_train.shape)
print("Testing Target Shape  :", y_test.shape)
```

### Observation

- The training dataset contains **1,176 employee records**, while the testing dataset contains **294 employee records**.
- The feature matrix consists of **44 predictor variables**, ensuring sufficient information for model training and evaluation.

### Business Interpretation

Using separate training and testing datasets allows the machine learning models to learn patterns from historical employee information while being evaluated on unseen employee records. Preserving the original attrition distribution through stratified sampling ensures a fair and reliable assessment of model performance.

## Feature Scaling Using StandardScaler

Machine learning algorithms often perform better when numerical features are measured on a similar scale. Features such as **Age**, **Monthly Income**, and **Total Working Years** have different ranges, which may influence model performance.

To standardize these numerical features, **StandardScaler** is applied. The scaler is fitted only on the training dataset and then used to transform both the training and testing datasets. This prevents data leakage and ensures a fair model evaluation.

### Identifying Continuous Numerical Features

Only continuous numerical features are selected for scaling. Binary variables generated through One-Hot Encoding are already represented as **0** and **1**, so they are left unchanged.

```python
# ============================================================
# Identifying Continuous Numerical Features
# ============================================================

continuous_features = [
    'Age',
    'DailyRate',
    'DistanceFromHome',
    'Education',
    'EnvironmentSatisfaction',
    'HourlyRate',
    'JobInvolvement',
    'JobLevel',
    'JobSatisfaction',
    'MonthlyIncome',
    'MonthlyRate',
    'NumCompaniesWorked',
    'PercentSalaryHike',
    'PerformanceRating',
    'RelationshipSatisfaction',
    'StockOptionLevel',
    'TotalWorkingYears',
    'TrainingTimesLastYear',
    'WorkLifeBalance',
    'YearsAtCompany',
    'YearsInCurrentRole',
    'YearsSinceLastPromotion',
    'YearsWithCurrManager'
]

print("Number of Continuous Features :", len(continuous_features))

print("\nContinuous Features:\n")
print(continuous_features)
```

### Applying StandardScaler

The scaler is fitted using only the training dataset and then applied to both the training and testing datasets. This approach prevents information from the testing dataset from influencing the training process, ensuring reliable model evaluation.

```python
# ============================================================
# Applying StandardScaler
# ============================================================

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[continuous_features] = scaler.fit_transform(
    X_train[continuous_features]
)

X_test[continuous_features] = scaler.transform(
    X_test[continuous_features]
)

print("Feature scaling completed successfully!")
```

### Verifying Feature Scaling

A sample of the scaled training dataset is displayed to verify that the selected numerical features have been successfully standardized while preserving the overall dataset structure.

## 4.1 Which Department Experiences the Highest Employee Attrition?

Understanding employee attrition across departments helps identify organizational units that experience higher turnover. This analysis enables Human Resource teams to prioritize departments requiring targeted retention strategies and workplace improvements.

```python
# ============================================================
# Attrition Rate by Department
# ============================================================

department_attrition = pd.crosstab(
    eda_data["Department"],
    eda_data["Attrition"],
    normalize="index"
) * 100

department_attrition = department_attrition.round(2)

department_attrition
```

```python
# ============================================================
# Visualization - Attrition Rate by Department
# ============================================================

department_attrition["Yes"].sort_values().plot(
    kind="barh",
    figsize=(8,5),
    color="steelblue"
)

plt.title("Employee Attrition Rate by Department")
plt.xlabel("Attrition Rate (%)")
plt.ylabel("Department")

plt.tight_layout()

plt.savefig("charts/01_Department_Attrition_Rate.png")

plt.show()
```

### Observation

- The **Sales** department recorded the highest employee attrition rate (**20.63%**), indicating that approximately one out of every five employees leaves the department.
- The **Human Resources** department showed the second-highest attrition rate (**19.05%**).
- **Research & Development** had the lowest attrition rate (**13.84%**), suggesting comparatively better employee retention.
- These results indicate that employee turnover varies across departments and may be influenced by department-specific workplace conditions.

### Business Interpretation

The analysis suggests that employee retention challenges are concentrated in specific departments rather than being evenly distributed across the organization. HR teams should prioritize retention strategies within the **Sales** and **Human Resources** departments by investigating workload, career growth opportunities, employee engagement, and workplace satisfaction. Targeted interventions in these departments could significantly reduce overall employee attrition.

# 5. Model Building & Comparison

Machine Learning models are developed to predict whether an employee is likely to leave the organization based on workplace, demographic, and job-related factors.

To ensure a reliable evaluation, three widely used classification algorithms are trained and compared. Each model is evaluated using multiple performance metrics rather than relying solely on accuracy, allowing the most suitable model for employee attrition prediction to be identified.

## 5.1 Business Objective

The objective of this stage is to develop a predictive machine learning model capable of identifying employees who are at risk of leaving the organization before attrition occurs. Early prediction enables Human Resource teams to implement proactive retention strategies, improve employee satisfaction, reduce recruitment costs, and strengthen workforce stability.

## 5.2 Model Selection Strategy

Three classification algorithms are selected for comparison based on their performance, interpretability, and practical applications in Human Resource Analytics.

- **Logistic Regression** serves as a simple and interpretable baseline model.
- **Random Forest Classifier** captures complex relationships and interactions among employee attributes.
- **Gradient Boosting Classifier** builds sequential decision trees to improve predictive performance and often achieves high classification accuracy.

Comparing these models helps identify the most effective approach for employee attrition prediction.

```python
# ============================================================
# Importing Additional Libraries for Model Building
# ============================================================

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

print("Model building libraries imported successfully!")
```

## 5.3 Logistic Regression

Logistic Regression is selected as the baseline classification model because it is simple, computationally efficient, and highly interpretable. It provides a useful benchmark for comparing the performance of more advanced machine learning algorithms.

```python
# ============================================================
# Logistic Regression Model
# ============================================================

logistic_model = LogisticRegression(
    class_weight="balanced",
    random_state=42,
    max_iter=1000
)

logistic_model.fit(X_train, y_train)

logistic_predictions = logistic_model.predict(X_test)

logistic_probabilities = logistic_model.predict_proba(X_test)[:,1]

print("Logistic Regression model trained successfully!")
```

## 5.4 Random Forest Classifier

Random Forest is an ensemble learning algorithm that combines multiple decision trees to improve prediction accuracy and reduce overfitting. It is widely used in HR Analytics because it effectively captures complex relationships among employee characteristics.

```python
# ============================================================
# Random Forest Classifier
# ============================================================

random_forest_model = RandomForestClassifier(
    n_estimators=200,
    class_weight="balanced",
    random_state=42
)

random_forest_model.fit(X_train, y_train)

rf_predictions = random_forest_model.predict(X_test)

rf_probabilities = random_forest_model.predict_proba(X_test)[:,1]

print("Random Forest model trained successfully!")
```

## 5.5 Gradient Boosting Classifier

Gradient Boosting builds decision trees sequentially, where each new tree attempts to correct the errors made by previous trees. This approach often provides strong predictive performance and is commonly applied to structured business datasets.

```python
# ============================================================
# Gradient Boosting Classifier
# ============================================================

gradient_boosting_model = GradientBoostingClassifier(
    random_state=42
)

gradient_boosting_model.fit(X_train, y_train)

gb_predictions = gradient_boosting_model.predict(X_test)

gb_probabilities = gradient_boosting_model.predict_proba(X_test)[:,1]

print("Gradient Boosting model trained successfully!")
```

## 5.6 Model Performance Comparison

The performance of all three classification models is evaluated using multiple evaluation metrics, including Accuracy, Precision, Recall, F1-Score, and ROC-AUC Score. Comparing multiple metrics provides a more comprehensive assessment than relying on accuracy alone, particularly for employee attrition prediction where identifying employees at risk is a critical business objective.

```python
# ============================================================
# Model Performance Comparison
# ============================================================

comparison = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],

    "Accuracy":[
        accuracy_score(y_test, logistic_predictions),
        accuracy_score(y_test, rf_predictions),
        accuracy_score(y_test, gb_predictions)
    ],

    "Precision":[
        precision_score(y_test, logistic_predictions),
        precision_score(y_test, rf_predictions),
        precision_score(y_test, gb_predictions)
    ],

    "Recall":[
        recall_score(y_test, logistic_predictions),
        recall_score(y_test, rf_predictions),
        recall_score(y_test, gb_predictions)
    ],

    "F1 Score":[
        f1_score(y_test, logistic_predictions),
        f1_score(y_test, rf_predictions),
        f1_score(y_test, gb_predictions)
    ],

    "ROC-AUC":[
        roc_auc_score(y_test, logistic_probabilities),
        roc_auc_score(y_test, rf_probabilities),
        roc_auc_score(y_test, gb_probabilities)
    ]

})

comparison = comparison.round(4)

comparison
```

### Model Comparison Summary

Although the Gradient Boosting model achieved the highest overall accuracy (85.03%), it identified only a small proportion of employees who actually left the organization. In contrast, the Logistic Regression model achieved the highest Recall (65.96%) and the highest ROC-AUC Score (0.8035), making it more effective at identifying employees who are at risk of attrition.

Since the primary objective of this project is early identification of employees who may leave the organization, **Logistic Regression is selected as the final model**. From an HR perspective, identifying more at-risk employees is more valuable than achieving slightly higher overall accuracy.

## 6.3 Receiver Operating Characteristic (ROC) Curve

The Receiver Operating Characteristic (ROC) Curve evaluates the ability of classification models to distinguish between employees who are likely to leave and those who are likely to stay. The Area Under the Curve (ROC-AUC) summarizes this performance into a single value, where a higher score indicates better classification capability.

```python
# ============================================================
# ROC Curve Comparison
# ============================================================

from sklearn.metrics import roc_curve, roc_auc_score

# ROC values
log_fpr, log_tpr, _ = roc_curve(y_test, logistic_probabilities)
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probabilities)
gb_fpr, gb_tpr, _ = roc_curve(y_test, gb_probabilities)

# ROC-AUC Scores
log_auc = roc_auc_score(y_test, logistic_probabilities)
rf_auc = roc_auc_score(y_test, rf_probabilities)
gb_auc = roc_auc_score(y_test, gb_probabilities)

plt.figure(figsize=(8,6))

plt.plot(
    log_fpr,
    log_tpr,
    label=f"Logistic Regression (AUC = {log_auc:.3f})"
)

plt.plot(
    rf_fpr,
    rf_tpr,
    label=f"Random Forest (AUC = {rf_auc:.3f})"
)

plt.plot(
    gb_fpr,
    gb_tpr,
    label=f"Gradient Boosting (AUC = {gb_auc:.3f})"
)

plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")

plt.legend()

plt.tight_layout()

plt.savefig("charts/08_ROC_Curve.png")

plt.show()

print(f"Logistic Regression ROC-AUC : {log_auc:.3f}")
print(f"Random Forest ROC-AUC      : {rf_auc:.3f}")
print(f"Gradient Boosting ROC-AUC  : {gb_auc:.3f}")
```

### Observation

- The ROC Curve comparison demonstrates that all three machine learning models perform better than random guessing.
- **Logistic Regression** achieved the highest ROC-AUC score (**0.804**), indicating the strongest ability to distinguish between employees who are likely to leave and those who are likely to stay.
- **Gradient Boosting** also performed well with an ROC-AUC of **0.794**, followed by **Random Forest** with **0.769**.
- Overall, Logistic Regression provided the best balance between predictive performance and model interpretability.

### Business Interpretation

From an HR analytics perspective, Logistic Regression provides the most suitable model for predicting employee attrition in this project. In addition to achieving the highest ROC-AUC score, the model is highly interpretable, enabling HR professionals to understand how different employee characteristics influence attrition risk. This transparency supports data-driven decision-making and helps organizations design targeted employee retention strategies.
